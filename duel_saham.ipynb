```python
import warnings
warnings.filterwarnings('ignore')
import yfinance as yf
import matplotlib.pyplot as plt
data_duel=yf.download(['AAPL','MSFT'],period='1mo',interval='1d')['Close']
print("Harga Penutupan Apple vs Microsoft:")
print(data_duel.head())
plt.figure(figsize=(12,6))
plt.plot(data_duel['AAPL'],label='Apple(AAPL)',color='blue')
plt.plot(data_duel['MSFT'],label='microsoft(MSFT)',color='orange')
plt.title('Perbandingan Harga Saham:Apple vs Microsoft')
plt.xlabel('Tanggal')
plt.ylabel('Harga(USD)')
plt.legend()
plt.grid(True)
plt.show()

data_normal=(data_duel/data_duel.iloc[0])*100
plt.figure(figsize=(12,6))
plt.plot(data_normal['AAPL'],label='Apple (AAPL)',color='blue')
plt.plot(data_normal['MSFT'],label='Microsoft (MSFT)',color='orange')
plt.axhline(y=100,color='black',linestyle='--', alpha=0.5)
plt.title('Perbandingan Performa: Siapa yang tumbuh lebih cepat?(Base 100)')
plt.xlabel('Tanggal')
plt.ylabel('Persentase Pertumbuhan(index 100)')
plt.legend()
plt.grid(True)
plt.show()
```
